```python
import pandas as pd
import re

# Load the file
with open("tfrc_prodigy.txt", "r") as f:
    content = f.read()

# Split into blocks
blocks = content.split("##########################################")
parsed_data = []

for block in blocks:
    if "Parsed structure file" not in block:
        continue

    data = {}
    m = re.search(r"Parsed structure file\s+(\S+)\s+\((\d+)\s+chains,\s+(\d+)\s+residues\)", block)
    if m:
        data['structure_id'] = m.group(1)
        data['chains'] = int(m.group(2))
        data['residues'] = int(m.group(3))

    fields = {
        "contacts_total": r"intermolecular contacts:\s+(\d+)",
        "charged_charged": r"charged-charged contacts:\s+([\d.]+)",
        "charged_polar": r"charged-polar contacts:\s+([\d.]+)",
        "charged_apolar": r"charged-apolar contacts:\s+([\d.]+)",
        "polar_polar": r"polar-polar contacts:\s+([\d.]+)",
        "apolar_polar": r"apolar-polar contacts:\s+([\d.]+)",
        "apolar_apolar": r"apolar-apolar contacts:\s+([\d.]+)",
        "apolar_percent": r"Percentage of apolar NIS residues:\s+([\d.]+)",
        "charged_percent": r"Percentage of charged NIS residues:\s+([\d.]+)",
        "binding_affinity_kcal_per_mol": r"binding\s+affinity\s+\(kcal\.mol-1\):\s*(-?[\d.]+)",
        "dissociation_constant_M": r"dissociation\s+constant\s+\(M\)\s+at\s+25\.0˚C:\s+([\deE\-.]+)"    
    }

    for key, regex in fields.items():
        m = re.search(regex, block)
        data[key] = float(m.group(1)) if m else None

    gap_matches = re.findall(r"^[HLT] \w+[\s\S]+?< Fragment \d+ > [HLT] \w+", block, re.MULTILINE)
    data["gap_count"] = len(gap_matches)

    parsed_data.append(data)

# Create DataFrame
df = pd.DataFrame(parsed_data)

# Optional: Sort by best binding affinity
df_sorted = df.sort_values(by="binding_affinity_kcal_per_mol")

# Display table
import IPython.display as display
display.display(df_sorted)
```
